​El Problema:
Recibes un array NumPy con características numéricas que están fuertemente afectadas por valores atípicos (outliers) y que poseen alta dimensionalidad. Debes mitigar la influencia de estos outliers y luego proyectar los datos a menos dimensiones para facilitar su visualización.
​Tu Misión: Escribe una función llamada pipeline_robusto_pca(X) que realice lo siguiente:
​Escale los datos utilizando RobustScaler de sklearn (para mitigar el impacto de los outliers basándose en la mediana y el rango intercuartílico).
​Aplique Análisis de Componentes Principales (PCA de sklearn) sobre los datos ya escalados para reducir la dimensionalidad a exactamente 2 componentes.
​Devuelva un array de numpy con la matriz transformada

```python
import numpy as np
from sklearn.preprocessing import RobustScaler
from sklearn.decomposition import PCA

def generar_caso_de_uso_pipeline_robusto_pca():
    # Generar dimensiones aleatorias
    n_samples = np.random.randint(50, 200)
    n_features = np.random.randint(5, 15)
    X = np.random.randn(n_samples, n_features) * 10

    # Añadir artificialmente outliers extremos
    outlier_indices = np.random.choice(n_samples, size=int(n_samples*0.1), replace=False)
    X[outlier_indices] = X[outlier_indices] * 100

    # Calcular salida esperada
    scaler = RobustScaler()
    X_scaled = scaler.fit_transform(X)
    pca = PCA(n_components=2)
    expected_output = pca.fit_transform(X_scaled)

    return {
        "input": {"X": X},
        "output": expected_output
    }
```

```python
generar_caso_de_uso_pipeline_robusto_pca()
```

​El Problema:
Tienes un DataFrame de pandas con registros transaccionales donde hay una columna de fechas en formato de texto (string). Necesitas extraer información temporal enriquecida para que un algoritmo de Machine Learning pueda capturar patrones estacionales.
​Tu Misión: Escribe una función llamada extraer_features_temporales(df, col_fecha) que realice lo siguiente sin alterar el DataFrame original:
​Cree una copia del DataFrame df.
​Convierta la columna indicada por col_fecha al tipo datetime de pandas.
​Extraiga el mes (1-12) en una nueva columna llamada mes.
​Extraiga el día de la semana (donde 0=Lunes y 6=Domingo) en una nueva columna llamada dia_semana.
​Cree una columna numérica (1 o 0) llamada es_fin_de_semana que sea 1 si el día es sábado (5) o domingo (6), y 0 en caso contrario.
​Elimine la columna original de texto col_fecha y devuelva el nuevo DataFrame modificado.

```python
import pandas as pd
import numpy as np

def generar_caso_de_uso_extraer_features_temporales():
    # Fechas aleatorias
    start_date = pd.to_datetime('2023-01-01')
    end_date = pd.to_datetime('2024-12-31')
    n_samples = np.random.randint(20, 100)

    random_days = np.random.randint(0, (end_date - start_date).days, n_samples)
    fechas = start_date + pd.to_timedelta(random_days, unit='d')

    df_input = pd.DataFrame({
        'fecha_str': fechas.astype(str),
        'ventas': np.random.uniform(100, 1000, n_samples)
    })

    # Output esperado
    df_output = df_input.copy()
    df_output['fecha_temp'] = pd.to_datetime(df_output['fecha_str'])
    df_output['mes'] = df_output['fecha_temp'].dt.month
    df_output['dia_semana'] = df_output['fecha_temp'].dt.dayofweek
    df_output['es_fin_de_semana'] = df_output['dia_semana'].isin([5, 6]).astype(int)
    df_output = df_output.drop(columns=['fecha_str', 'fecha_temp'])

    return {
        "input": {"df": df_input.copy(), "col_fecha": 'fecha_str'},
        "output": df_output
    }
```

​El Problema:
Te enfrentas a un problema de clasificación en el que quieres evaluar un modelo K-Nearest Neighbors (KNN). Sin embargo, quieres darle más peso a los vecinos más cercanos en lugar de que todos los "k" vecinos voten por igual.
​Tu Misión: Escribe una función evaluar_knn_pesos(X_train, y_train, X_test, n_vecinos) que realice lo siguiente:
​Instancie un modelo KNeighborsClassifier de sklearn utilizando el número de vecinos proporcionado (n_vecinos) y asegurándose de usar el parámetro weights='distance' para la ponderación por distancia.
​Entrene (ajuste) el modelo utilizando los datos de entrenamiento (X_train, y_train).
​Devuelva un array de numpy con las etiquetas predichas para el conjunto de prueba (X_test).

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_classification

def generar_caso_de_uso_evaluar_knn_pesos():
    X, y = make_classification(
        n_samples=np.random.randint(50, 150),
        n_features=np.random.randint(4, 10),
        random_state=np.random.randint(0, 1000)
    )

    split_idx = int(len(X) * 0.8)
    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train = y[:split_idx]
    n_vecinos = np.random.randint(3, 8)

    # Lógica esperada
    clf = KNeighborsClassifier(n_neighbors=n_vecinos, weights='distance')
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    return {
        "input": {
            "X_train": X_train,
            "y_train": y_train,
            "X_test": X_test,
            "n_vecinos": n_vecinos
        },
        "output": y_pred
    }
```

El Problema:
Tienes un array de variables numéricas fuertemente sesgadas a la derecha. Quieres suavizar la distribución aplicando una transformación matemática antes de agrupar los datos en clústeres.

Tu Misión:
Escribe una función log_transform_kmeans(X, n_clusters) que realice lo siguiente:
- Aplique una transformación logarítmica usando np.log1p (de numpy) a toda la matriz X.
- Instancie y entrene un modelo KMeans de sklearn sobre los datos transformados usando el n_clusters indicado y random_state=42.
- Devuelva un diccionario con dos claves: 'labels' (el array de etiquetas asignadas) y 'centers' (los centros de los clústeres en escala logarítmica).


```python
import numpy as np
from sklearn.cluster import KMeans

def generar_caso_de_uso_log_transform_kmeans():
    n_samples = np.random.randint(50, 200)
    n_features = np.random.randint(2, 5)
    # Datos con sesgo a la derecha
    X = np.random.exponential(scale=100.0, size=(n_samples, n_features))
    n_clusters = np.random.randint(2, 6)

    # Lógica esperada
    X_log = np.log1p(X)
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto')
    kmeans.fit(X_log)

    expected_output = {
        'labels': kmeans.labels_,
        'centers': kmeans.cluster_centers_
    }

    return {
        "input": {"X": X, "n_clusters": n_clusters},
        "output": expected_output
    }
```
